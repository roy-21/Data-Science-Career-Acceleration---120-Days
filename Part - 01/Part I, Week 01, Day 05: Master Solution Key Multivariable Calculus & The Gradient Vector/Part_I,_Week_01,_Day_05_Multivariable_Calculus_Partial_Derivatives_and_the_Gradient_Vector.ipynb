```python
import numpy as np
import sympy as sp

# ==========================================
# 1. Symbolic Partial Derivatives (Exact Math)
# ==========================================

# Define symbolic variables
x, y = sp.symbols('x y')

# Define the multivariable function: f(x, y) = x^2 * y + sin(x) + e^y
f = (x**2) * y + sp.sin(x) + sp.exp(y)

# Calculate partial derivatives symbolically
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("--- Symbolic Partial Derivatives (SymPy) ---")
print(f"Function f(x,y) = {f}")
print(f"df/dx = {df_dx}")
print(f"df/dy = {df_dy}\n")

# Evaluate the gradient vector at a specific point (x=1, y=2)
grad_at_point = np.array([
    float(df_dx.subs({x: 1, y: 2})),
    float(df_dy.subs({x: 1, y: 2}))
])

print(f"Gradient Vector at (1, 2): {grad_at_point}\n")
```

```python
# ==========================================
# 2. Numerical Gradient Approximation
# ==========================================

# Definition of derivative: (f(x+h) - f(x)) / h
def numerical_partial_derivative(func, point, var_index, h=1e-5):
    """
    Approximates the partial derivative of a function at a specific point.
    var_index: 0 for x, 1 for y
    """
    point_shifted = np.array(point, dtype=float)
    point_shifted[var_index] += h

    # Rise over run
    return (func(*point_shifted) - func(*point)) / h


# Define the same function using NumPy
def f_num(x_val, y_val):
    return (x_val**2) * y_val + np.sin(x_val) + np.exp(y_val)


point = [1.0, 2.0]

approx_df_dx = numerical_partial_derivative(f_num, point, 0)
approx_df_dy = numerical_partial_derivative(f_num, point, 1)

approx_gradient = np.array([approx_df_dx, approx_df_dy])

print("--- Numerical Gradient Approximation (NumPy) ---")
print(f"Approximated Gradient at (1, 2): {approx_gradient}")
print("Notice how closely the numerical approximation matches the exact symbolic math!")
```
