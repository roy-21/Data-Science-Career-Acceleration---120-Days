```python
import sympy as sp
import numpy as np

# Problem 1: The Jacobian of a Vector-Valued Function
x1, x2 = sp.symbols('x1 x2')

y1 = x1**2 + sp.sin(x2)
y2 = x1 * x2
y3 = sp.exp(x1) + 2 * x2

Y = sp.Matrix([y1, y2, y3])
X = sp.Matrix([x1, x2])

# Compute the symbolic Jacobian matrix
Jacobian_sym = Y.jacobian(X)

# Evaluate the Jacobian at the point (x1 = 1, x2 = 0)
Jacobian_num = Jacobian_sym.subs({x1: 1, x2: 0})

print("Problem 1 Results:")
print(f"Symbolic Jacobian:\n{Jacobian_sym}\n")
print(f"Evaluated Jacobian at (1, 0):\n{np.array(Jacobian_num).astype(float)}\n")
```

```python
# Problem 2: The Hessian and Curvature Classification
def classify_critical_point(hessian_matrix):
    """
    Takes a NumPy array representing a Hessian matrix
    and returns its classification based on eigenvalues.
    """

    # Compute the eigenvalues
    eigenvalues, _ = np.linalg.eig(hessian_matrix)

    # Classification logic
    # Small tolerance for floating-point comparisons
    if np.all(eigenvalues > 1e-6):
        classification = "Local Minimum"
    elif np.all(eigenvalues < -1e-6):
        classification = "Local Maximum"
    else:
        classification = "Saddle Point"

    return eigenvalues, classification


H_A = np.array([
    [4, 1],
    [1, 3]
])

H_B = np.array([
    [-2, 0],
    [0, -5]
])

H_C = np.array([
    [2, 4],
    [4, -1]
])

matrices = {
    "Point A": H_A,
    "Point B": H_B,
    "Point C": H_C
}

print("Problem 2 Results:")
for name, H in matrices.items():
    evals, result = classify_critical_point(H)
    print(f"{name}: Eigenvalues {evals.round(2)} -> {result}")
```
