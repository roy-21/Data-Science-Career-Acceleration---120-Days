```python

import numpy as np

# =====================================
# Problem 1: Accelerating Matrix Powers
# =====================================

A = np.array([
    [0.8, 0.3],
    [0.2, 0.7]
])

k = 10

# Method 1: Direct Power
A_k_direct = np.linalg.matrix_power(A, k)

# Method 2: Eigendecomposition
eigenvalues, Q = np.linalg.eig(A)

# Create diagonal matrix Lambda^k
Lambda_k = np.diag(eigenvalues ** k)

# Reconstruct A^k
Q_inv = np.linalg.inv(Q)
A_k_eigen = Q @ Lambda_k @ Q_inv

print("Problem 1 Results:")
print(f"Direct Power:\n{A_k_direct}\n")
print(f"Eigendecomposition Power:\n{A_k_eigen}\n")


# =====================================
# Problem 2: The Core of PCA (1D Projection)
# =====================================

np.random.seed(42)

X = np.random.randn(100, 2) @ np.array([
    [2, 1],
    [1, 2]
])

# Covariance matrix
cov_matrix = np.cov(X, rowvar=False)

# Eigen decomposition
evals, evecs = np.linalg.eig(cov_matrix)

# Principal eigenvector (largest eigenvalue)
principal_idx = np.argmax(evals)
principal_eigenvector = evecs[:, principal_idx]

# Project data to 1D
X_projected = X @ principal_eigenvector

print("Problem 2 Results:")
print(f"Covariance Matrix:\n{cov_matrix}\n")
print(f"Principal Eigenvector: {principal_eigenvector}")
print(f"Shape of compressed data: {X_projected.shape}\n")


# =====================================
# Problem 3: The Symmetry Superpower
# =====================================

S = np.array([
    [3, 1],
    [1, 3]
])

# Eigen decomposition
sym_evals, sym_evecs = np.linalg.eig(S)

# Check orthogonality
orthogonality_check = np.dot(sym_evecs[:, 0], sym_evecs[:, 1])

print("Problem 3 Results:")
print(f"Dot product of eigenvectors: {orthogonality_check:.10f}")
print("If the result is ~0, they are perfectly orthogonal.")
```
